**Creation of Dataframes using pandas!**

```python
import numpy as np
import pandas as pd
```

```python
#Creating our own dataframe
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"])
```

```python
df.head()
```

```python
#accesing the index
df.index
```

```python
#accesing the columns
df.columns
```

```python
#we can also make index non numeric
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"],index=["x","y","z"])
```

```python
#for info about the data
df.info()
```
