**Creation of Dataframes using pandas!**

```python
import numpy as np
import pandas as pd
```

```python
#using lists
student_data=[
    [100,80,70],
    [50,60,70],
    [60,70,80],
]
pd.DataFrame(student_data,columns=['Maths','Science','English'])
```

```python
#using dicts
student_data={
    'Maths':[100,50,60],
    'Science':[80,60,70],
    'English':[70,70,80],
}
pd.DataFrame(student_data)
```
