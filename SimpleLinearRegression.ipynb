```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/placement (3).csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
#step 1 separate input and output columns
X=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
#visualising the line
plt.scatter(df['cgpa'],df['package'])
plt.plot(df['cgpa'],lr.predict(X))
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
#after training our linear model has some attribute
m=lr.coef_
```

```python
b=lr.intercept_
```
